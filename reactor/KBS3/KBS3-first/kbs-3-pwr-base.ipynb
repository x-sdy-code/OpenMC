```python
Pin_Cu_inR6 = openmc.ZCylinder(r= 94.9 /2)
Pin_Cu_outR3 = openmc.ZCylinder(r= 105 /2,boundary_type="vacuum")
pin_Cu_up1 = openmc.ZPlane(z0= 241.75)
pin_Cu_up2 = openmc.ZPlane(z0= 236.75)
pin_Cu_up3 = openmc.ZPlane(z0= 230)

pin_Cu_down1 = openmc.ZPlane(z0= -241.75)
pin_Cu_down2 = openmc.ZPlane(z0= -236.75)
pin_Cu_down3 = openmc.ZPlane(z0= -230)

clad_Cu= openmc.Cell(name="clad_Cu")
clad_Cu.fill=Cu_m1
clad_Cu.region = -Pin_Cu_outR3 &+ Pin_Cu_inR6 &+pin_Cu_down1 &-pin_Cu_up1

clad_up_Cu = openmc.Cell(name="clad_up_Cu")
clad_up_Cu.fill=Cu_m1
clad_up_Cu.region = -Pin_Cu_inR6  &+pin_Cu_up3 &-pin_Cu_up2

clad_down_Cu = openmc.Cell(name="clad_down_Cu")
clad_down_Cu.fill=Cu_m1
clad_down_Cu.region = -Pin_Cu_inR6  &+pin_Cu_down2 &-pin_Cu_down3

clad_Fe_up = openmc.Cell(name="clad_Fe_up")
clad_Fe_up.fill=Fe_m7
clad_Fe_up.region = -Pin_Cu_inR6 &+pin_main_up22 &-pin_Cu_up3

clad_Fe_down = openmc.Cell(name="clad_Fe_down")
clad_Fe_down.fill=Fe_m7
clad_Fe_down.region = -Pin_Cu_inR6 &+pin_Cu_down3 &-pin_main_down23


# clad_Cu_U = openmc.Universe(cells=[clad_Cu, clad_up_Cu,clad_down_Cu,
#                                    clad_Fe_up])
# clad_Cu_U.plot(width=(200.,805), basis='xz',pixels=(400,400),origin=(0,23,0),color_by = 'material')
```

```python
Room_R = 175/2
Room_down = -400
Room_up = 400

Room_R = openmc.ZCylinder(r=Room_R,boundary_type='vacuum')
Room_up = openmc.ZPlane(Room_up,boundary_type='vacuum')
Room_down = openmc.ZPlane(Room_down,boundary_type='vacuum')

Room_cell =  openmc.Cell(fill=PRT_m9,name="vacuum_world_cell")
Room_cell.region = -Room_R &+Room_down &-Room_up
```

```python
KBS = openmc.Universe(name='KBS Lattice')
core = openmc.Cell(name='core cell', fill=assembly_235fuel_u1)
core.region =  -Pin_Cu_inR6 &+pin_main_down23 &-pin_main_up22

Room_cell.region &~ core.region
Room_cell.region &~ clad_Cu.region
Room_cell.region &~ clad_up_Cu.region
Room_cell.region &~ clad_down_Cu.region
Room_cell.region &~ clad_Fe_up.region
Room_cell.region &~ clad_Fe_down.region

KBS.add_cells([clad_Cu, clad_up_Cu,clad_down_Cu,clad_Fe_up,clad_Fe_down,
               core,Room_cell])
geometry = openmc.Geometry(KBS)
geometry.export_to_xml()
```
